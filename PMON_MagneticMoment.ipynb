```python
import numpy as np
import matplotlib.pyplot as plt

# We calcualte the total magnetic moment (assuming 1.6 Bohr magnetron, units in Am^2):
BU=9.27e-24;
#Moment=NumAt*1.6*BU, units of Am^2
Moment=0.001
mu=4.*np.pi*1e-7
x_TM = np.linspace(0.1, 2.55, 50)
y_TM = np.linspace(0.1, 2.55, 50)
# distance in cm
dist_TM = np.sqrt(x_TM**2+y_TM**2)*30
# we calculate the feild using, B=mu*moment/2*pi*R^3, in nT
field=mu*Moment/(2.*np.pi*(dist_TM/100)**3)*1e9
field_1m=mu*Moment/(2.*np.pi*1**3)*1e9
field_50cm=mu*Moment/(2.*np.pi*0.5**3)*1e9
field_25cm=mu*Moment/(2.*np.pi*0.25**3)*1e9
print("Moment (Ampere*m^2) = ", Moment)
print("Magnetic Field at 25cm (nT) = ", field_25cm)
print("Magnetic Field at 50cm (nT) = ", field_50cm)
print("Magnetic Field at 1m (nT) = ", field_1m)
plt.style.use('bmh')
plt.plot(dist_TM, field, 'g-')
plt.plot(dist_TM,np.ones(np.shape(dist_TM))*500,'r-')
plt.plot(dist_TM,np.ones(np.shape(dist_TM))*12.8,'k-.')
plt.axvline(x = 25)
plt.yscale("log")
# plt.xscale("log")
plt.title("Field vs Distance to TM for magnetic moment of 1mAm^2")
plt.xlabel("Radial Distance to TM (cm)")
plt.ylabel("Magnetic Field (nT)")
plt.show()
```
